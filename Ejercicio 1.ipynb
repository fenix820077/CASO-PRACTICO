Para procesar el archivo hay que tener en cuenta que cada palabra o conjunto de palabras están separadas de su valor mediante el carácter “\t”. 

La solucion propuesta consta de los siguientes códigos ó funciones:

·def cargar_sentimientos(Sentimientos):

·def procesar_json(Tweets):

·def ejercicio_1(Sentimientos, Tweets):
Esta es la funcion principal del ejercicio y llama a las dos funciones anteriores.

#DESARROLLO PARTE 2
Para desarrollar la segunda parte, debemos iportar una libreria poderosa denominada nltk
Definimos datos indicando la dirección de los tweets y verificamos si lee los datos del archivo usando 



```python
#importar modulos
import json

#Funcion que lee el archivo de sentimientos y devuelve un diccionario cargado con los colores int de los sentimientos
def cargar_sentimientos(file1):
    sentimientos=open(file1)
    valores={}
    for linea in sentimientos:
        termino,valor=linea.split("\t")
        valores[termino]= int(valor)
        
    return valores


#Procesado del fichero json que devuelve un array de diccionarios (uno por tweet)
def procesar_json(file):
    paraprocesar=open(file,'r')
    cadena=paraprocesar.read()
    paraprocesar.close()

    replaced=cadena.replace("}\n{", "},{") #reemplazando "}/n{" por "}{"
    replaced='['+replaced+']'              #abriendo y cerrando con corchetes para convertir la cadena en un array de diccionarios
    
    output= open('Tweets.json','w')
    output.write(replaced)
    output.close()
    #Carga de los tweets del fichero
    with open('Tweets.json','r')as file:
        try: tweets=json.load(file)
        except:  tweets=None
    #  Carga finalizada
    return tweets #Devuelve un array de diccionarios

#Funcion ppal ejercicio 1: Como parametros de entrada tiene el archivo de sentimientos y el archivo de tweets,
#y como resultado imprime por pantalla el  valor sentimental calculado.
def ejercicio_1(file_sent, file_tweets): 
    #Carga del diccionario de entimientos
    diccionario_sentimental=cargar_sentimientos(file_sent)
    k="text"
    sent_claves=diccionario_sentimental.keys()
    
    #Carga de los tweets del fichero
    tweets=procesar_json(file_tweets)
    
    #Procesado de los tweets uno a uno
   
    for x in range(0,len(tweets)):  #recorriendo el array de tweets
        if k in tweets[x].keys(): #si el tweet tiene texto
            texto=tweets[x].get(k).lower()#asigno el texto del tweet a una cadena
            palabras=texto.split()
            #Para cada palabra del tweet, miro si esta en el diccionario de sentimientos, y sumo su valor
            valor_tweet=0
            for sentimiento in palabras: 
                if sentimiento in sent_claves:
                    valor_sentimiento=int(diccionario_sentimental.get(sentimiento))
                    valor_tweet+=valor_sentimiento
                else: pass
            print("El siguiente tweet, con id "+ str(tweets[x].get('id'))+' "'+texto+'"')
            print("Tiene un sentimiento de: " + str(valor_tweet)+"\n")
    
        else: pass  
        

#Peticion de parametros de entrada y llamada al ejercicio:

print("Escriba el archivo que contiene los SENTIMIENTOS")
file_dic=input()
print("ESCRIBA EL NOMBRE DEL ARCHIVO QUE CONTIENE TWEETS")
file_tweets=input()
print("\n")
ejercicio_1(file_dic,file_tweets)
```
